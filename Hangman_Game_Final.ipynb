## **Created the Txt File with Words.**

```python
with open('words.txt','a',newline='') as f:
  word = input('Enter Any Word: ')
  f.write(word + '\n')

  while True:

    option = input('Do you want to enter new word: [Y | N] : ')

    if option.lower() == 'y':
      word = input('Enter Any Word: ')
      f.write(word + '\n')

    elif option.lower() == 'n':
      print('Data Written Successfully....')
      break
```

## **Business Logic For Game**

```python
def Hungman_Game():
    
    while True:
        # Reading the text file and storing it in list

        word_list = []
        final_list = []

        with open('words.txt','r') as f:
            read_obj = f.readlines()

            for line in read_obj:
                word_list.append(line)

        for x in word_list:
            final_list.append(x.replace('\n',''))


        # Getting the random words from the list

        import random

        word = random.choice(final_list)
        # print(word)
        # print()

        # Converting the wors into tokens

        letters = []

        for x in word:
            letters.append(x)    

        # Getting the count of words

        letters_count = len(letters)    

        # Printing the dashes

        dashes = '_'*letters_count
        empty_dash = []
        for x in dashes:
            empty_dash.append(x)    

        # Printing the empty dashes

        for x in empty_dash:
            print(x,end=' ')
        print()
        print()

        count = 0

        while True:       

            print()
            guess_letter = input('Guess the letter: ')
            print()

            # Getting the index of the guessed letter.

            import re

            match = re.finditer(guess_letter,word)

            indexes = []       

            for x in match:

                if x is not None:
                    indexes.append(x.start())
                    empty_dash[x.start()] = x.group()

            print(' '.join(empty_dash))            

            if len(indexes) == 0:
                print()
                print('Wrong Guess..')
                print()
                print('You have {} chances to guess the word...!!!'.format(letters_count - count))
                print()
                count = count+1 

            final_empty_dash = ''.join(empty_dash)              

            if word == final_empty_dash:
                print()
                print('Congratulations you have guessed the right word...!!!')
                break
                print()

            if count > letters_count:           

                print('#'*50)
                print('\t      Word =',word)
                print()
                print('\t\tGame Over')
                print()
                print('\t\tPlay Again')
                print('#'*50)
                break
                
        print()
        option = input('Do you want to play again [Y|N]: ')
        print('#'*50)
        print()
        
        
        if option.lower() == 'n':
            print('Thanks for playing the game...!!!')
            print()
            break

Hungman_Game()
```
